```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
# Se ajusta el tipo de dato

df['release_year'] = df['release_year'].astype(int)
df['main_country'] = df['country'].fillna('Unknown').str.split(',').str[0].str.strip()
df['type'] = df['type'].astype(str)

df['duration'] = df['duration'].astype(str)
df['duration_value'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract(r'(\D+)$').astype(str).apply(lambda x: x.str.strip())
```

```python
df.info()
```

```python
# Introducimos los Códigos para las paletas de color

# Monocromática roja
netflix_mono = ["#8A0008", "#B20710", "#E50914", "#FF4A55", "#FF7A82", "#FFB4B7"]

# Complementaria (rojo + aqua)
netflix_complementary = ["#E50914", "#09E5DA", "#079C94", "#73FFF6"]

# Triádica
netflix_triadic = ["#E50914", "#1409E5", "#09E514"]

# Tétrada
netflix_tetrad = ["#E50914", "#09E5DA", "#E5DA09", "#5A09E5"]

# Paleta clásica Netflix (rojo + negros/grises)
netflix_classic = ["#E50914", "#221F1F", "#666666", "#B3B3B3", "#FFFFFF"]

# Escalas continuas tipo Netflix para heatmap/mapa
netflix_continuous_red = [
    (0.0, "#221F1F"),
    (0.5, "#E50914"),
    (1.0, "#FFB4B7")]

netflix_continuous_map = [
    (0.0, "#221F1F"),
    (0.5, "#E50914"),
    (1.0, "#FFE5E7")]

# Estilo base de layout estilo Netflix oscuro
base_layout = dict(
    template='plotly_dark',
    paper_bgcolor='#000000',
    plot_bgcolor='#000000',
    font=dict(color='#FFFFFF'))
```

```python
# Primera gráfica. Conteo de películas.

fig1_data = df['type'].value_counts().reset_index()
fig1_data.columns = ['type', 'count']

fig1 = px.bar(
    fig1_data,
    x='type', y='count',
    color='type',
    color_discrete_sequence=netflix_mono,
    text='count',
    title='Distribución de títulos por tipo (Movie vs TV Show)',
    labels={'type': 'Tipo', 'count': 'Cantidad de títulos'}
)
fig1.update_traces(
    hovertemplate='Tipo: %{x}<br>Títulos: %{y}',
    textposition='outside'
)
fig1.update_layout(
    **base_layout,
    uniformtext_minsize=12,
    uniformtext_mode='hide'
)
fig1.show()

print("""
Gráfico 1 — Conteo de Películas vs Series
Este gráfico muestra cuántos títulos en el catálogo de Netflix corresponden a películas y cuántos a series.
Permite ver rápidamente qué tipo de contenido domina la plataforma y comparar la magnitud entre ambas categorías.
""")
```

```python
# Segunda gráfica. Titulos lanzados por año.

titles_per_year = df.groupby('release_year').size().reset_index(name='count')
titles_per_year = titles_per_year.sort_values('release_year')

fig2 = px.line(
    titles_per_year,
    x='release_year', y='count',
    markers=True,
    title='Número de títulos lanzados por año',
    labels={'release_year': 'Año de lanzamiento', 'count': 'Cantidad de títulos'}
)
fig2.update_traces(
    line=dict(color="#E50914", width=3),
    marker=dict(size=6, color="#FF4A55"),
    hovertemplate='Año: %{x}<br>Títulos: %{y}'
)
fig2.update_layout(**base_layout)
fig2.show()

print("""
Gráfico 2 — Evolución de títulos por año de lanzamiento
Este gráfico muestra cómo ha cambiado el número de títulos disponibles en Netflix según su año de lanzamiento.
Se pueden identificar tendencias de crecimiento, años con más producción y épocas con menos contenidos lanzados.
""")
```

```python
# Tercera gráfica. Top 10 países con más titulos.
top_countries = (
    df['main_country']
    .value_counts()
    .head(10)
    .reset_index()
)
top_countries.columns = ['country', 'count']

fig3 = px.bar(
    top_countries,
    x='count', y='country',
    color='country',
    color_discrete_sequence=netflix_complementary,
    orientation='h',
    title='Top 10 países con más títulos en Netflix',
    labels={'country': 'País', 'count': 'Cantidad de títulos'},
    hover_data={'country': True, 'count': True}
)
fig3.update_layout(
    **base_layout,
    yaxis={'categoryorder': 'total ascending'}
)
fig3.update_traces(
    hovertemplate='País: %{y}<br>Títulos: %{x}'
)
fig3.show()

print("""
Gráfico 3 — Top 10 países con más títulos
Aquí se observan los diez países que más títulos aportan al catálogo de Netflix.
Permite identificar qué regiones son las más representadas en la plataforma y comparar el peso relativo de cada país.
""")
```

```python
# Cuarta gráfica. Distribución del rating.

ratings = df['rating'].fillna('Unknown').value_counts().reset_index()
ratings.columns = ['rating', 'count']

fig4 = px.bar(
    ratings,
    x='rating', y='count',
    color='rating',
    color_discrete_sequence=netflix_triadic + netflix_mono,
    title='Distribución de clasificaciones de contenido (Rating)',
    labels={'rating': 'Clasificación', 'count': 'Cantidad de títulos'},
    hover_data={'rating': True, 'count': True}
)
fig4.update_traces(
    hovertemplate='Clasificación: %{x}<br>Títulos: %{y}'
)
fig4.update_layout(
    **base_layout,
    xaxis_tickangle=-45
)
fig4.show()

print("""
Gráfico 4 — Distribución de clasificaciones de contenido
Este gráfico resume cuántos títulos existen en cada clasificación por edad (por ejemplo, TV-MA, TV-14, PG, etc.).
Ayuda a entender hacia qué tipo de público está más orientado el contenido disponible en Netflix.
""")
```

```python
# Quinta gráfica. Distribución de generos y categorías en Netflix.

genres = (
    df.assign(
        genre=df['listed_in'].fillna('Unknown').str.split(', ')
    )
    .explode('genre')
)

genres_count = genres['genre'].value_counts().reset_index()
genres_count.columns = ['genre', 'count']

fig5 = px.treemap(
    genres_count,
    path=['genre'],
    values='count',
    title='Distribución de géneros/categorías en Netflix',
    hover_data={'count': True},
    color='count',
    color_continuous_scale=netflix_continuous_red
)
fig5.update_traces(
    hovertemplate='Género: %{label}<br>Títulos: %{value}'
)
fig5.update_layout(**base_layout)
fig5.show()

print("""
Gráfico 5 — Treemap de géneros y categorías
El treemap muestra cómo se distribuyen los títulos según su género o categoría principal.
Cada bloque representa un género y su tamaño refleja la cantidad de títulos, lo que permite ver cuáles son los más predominantes.
""")
```

```python
# Septima gráfica. Duración promedio de películas

movies = df[df['type'] == 'Movie'].copy()
movies = movies[movies['duration_unit'].str.contains('min', case=False, na=False)]

fig6 = px.box(
    movies,
    y='duration_value',
    points='all',
    title='Distribución de duración de las películas (minutos)',
    labels={'duration_value': 'Duración (min)'}
)
fig6.update_traces(
    marker=dict(size=4, color="#FF4A55", opacity=0.7),
    line=dict(color="#E50914"),
    hovertemplate='Duración: %{y} min'
)
fig6.update_layout(**base_layout)
fig6.show()

print("""
Gráfico 6 — Boxplot de duración de películas
Este boxplot representa la distribución de la duración de las películas en minutos, mostrando mediana, cuartiles y posibles valores atípicos.
Ayuda a identificar si la mayoría de las películas tienen duraciones similares o si existe mucha variabilidad.
""")
```

```python
# Séptima gráfica. Duración promedio de películas por década.

movies['decade'] = (movies['release_year'] // 10) * 10

avg_duration_decade = (
    movies.groupby('decade')['duration_value']
    .mean()
    .reset_index()
)

fig7 = px.bar(
    avg_duration_decade,
    x='decade', y='duration_value',
    text='duration_value',
    title='Duración promedio de películas por década',
    labels={'decade': 'Década', 'duration_value': 'Duración promedio (min)'},
    color='decade',
    color_continuous_scale=netflix_continuous_red
)
fig7.update_traces(
    hovertemplate='Década: %{x}s<br>Duración promedio: %{y:.1f} min',
    texttemplate='%{y:.1f}',
    textposition='outside'
)
fig7.update_layout(**base_layout)
fig7.show()

print("""
Gráfico 7 — Duración promedio por década
En este gráfico se observa cómo ha evolucionado la duración promedio de las películas a lo largo de diferentes décadas.
Permite detectar si las películas tienden a hacerse más largas, más cortas o si mantienen una duración relativamente estable.
""")
```

```python
# Octava gráfica. Mapa mundial de títulos por país.

country_counts = (
    df['main_country']
    .value_counts()
    .reset_index()
)
country_counts.columns = ['country', 'count']

fig8 = px.choropleth(
    country_counts,
    locations='country',
    locationmode='country names',
    color='count',
    color_continuous_scale=netflix_continuous_map,
    title='Distribución mundial de títulos de Netflix por país principal',
    labels={'count': 'Cantidad de títulos'},
    hover_data={'country': True, 'count': True}
)
fig8.update_traces(
    hovertemplate='País: %{location}<br>Títulos: %{z}'
)
fig8.update_layout(**base_layout)
fig8.show()

print("""
Gráfico 8 — Mapa mundial de títulos por país
Este mapa muestra la cantidad de títulos asociados a cada país principal en el catálogo de Netflix.
Los países con tonalidades más claras cuentan con mayor número de títulos, lo que facilita identificar concentraciones geográficas de contenido.
""")
```

```python
# Novena gráfica. Cantidad de títulos agregados por mes.

# Convertir la fecha a datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extraer mes
df['added_month'] = df['date_added'].dt.month

# Contar títulos por mes
added_month_count = df.groupby('added_month').size().reset_index(name='count')

fig9 = px.bar(
    added_month_count,
    x='added_month',
    y='count',
    color='added_month',
    color_continuous_scale=netflix_continuous_red,
    title='Cantidad de títulos añadidos por mes',
    labels={'added_month': 'Mes', 'count': 'Cantidad de títulos'}
)

fig9.update_traces(
    hovertemplate='Mes: %{x}<br>Títulos añadidos: %{y}',
    marker=dict(line=dict(color='#FFFFFF', width=1))
)

fig9.update_layout(
    **base_layout,
    xaxis=dict(dtick=1)
)

fig9.show()

print("""
Gráfico 9 — Títulos añadidos por mes
Este gráfico muestra en qué meses Netflix agrega más contenido nuevo a la plataforma.
Refleja patrones estacionales de actualización del catálogo y permite identificar picos de carga de contenido.
""")
```

```python
# Decima gráfica. Duración promedio de peliculas por clasificación.

duration_rating = (
    df[df['duration_value'].notnull()]
    .groupby(['rating', 'type'])['duration_value']
    .mean()
    .reset_index()
)

fig10 = px.treemap(
    duration_rating,
    path=['rating', 'type'],
    values='duration_value',
    title='Duración promedio del contenido por clasificación y tipo',
    color='duration_value',
    color_continuous_scale=netflix_continuous_red,
    labels={'duration_value': 'Duración promedio (min)'}
)

fig10.update_traces(
    hovertemplate=
        'Clasificación: %{label}<br>' +
        'Tipo: %{parent}<br>' +
        'Duración promedio: %{value:.1f} min'
)

fig10.update_layout(**base_layout)
fig10.show()

print("""
Gráfico 10 — Duración promedio por clasificación y tipo
Este treemap muestra qué clasificaciones tienen contenidos más largos en promedio y cómo se divide esa duración entre películas.
Los bloques más grandes representan clasificaciones con mayor duración promedio.
""")
```
